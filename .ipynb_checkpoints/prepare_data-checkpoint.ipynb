```python
import pandas as pd
import numpy as np
import glob
import os
from pathlib import Path


start_year = 2017

# adds dir
odds_path = './odds/' 
tml_path = './TML-Database/'
players_atp_path = "./TML-Database/ATP_Database.csv"
```

```python
excel_files = glob.glob(os.path.join(odds_path, '*.xlsx'))
excel_files =  list(filter(
    lambda f: int(Path(f).stem.strip()) >= start_year,
    excel_files
))
df_list = []

for file in excel_files:
    df = pd.read_excel(file)
    df_list.append(df)

odds_df = pd.concat(df_list, ignore_index=True)
odds_df = odds_df.replace('-', np.nan)
odds_df.sample(8).T
```

```python
odds_df.to_parquet('./ods.parquet')
```

```python
# odds_df.sample(4).to_string()
```

```python
csv_files = glob.glob(os.path.join(tml_path, '*.csv'))
csv_files =  sorted(
    filter(lambda f: Path(f).stem.strip().isdigit() and int(Path(f).stem.strip()) >= start_year, csv_files),
    key=lambda f: int(Path(f).stem.strip())
)
df_list = []

for file in csv_files:
    df = pd.read_csv(file)
    df_list.append(df)

tml_df = pd.concat(df_list, ignore_index=True)
tml_df = tml_df.replace('-', np.nan)
tml_df.sample(8).T
```

```python
tml_df.to_parquet('./tml.parquet')
```

```python
# tml_df.sample(4).to_string()
```

```python
pls_df = pd.read_csv(players_atp_path, encoding='latin1')
```

```python
pls_df.sample(8).T
```

```python
pls_df.to_parquet('./players.parquet')
```
