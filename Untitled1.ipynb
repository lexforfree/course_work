```python
import numpy as np
from statsmodels.stats.proportion import proportions_ztest

# Данные
counts = np.array([2_520, 2592])  # успехи
# counts = np.array([2_520, 2592])  # успехи

nobs = np.array([9000, 9000])    # размеры выборок
nobs = np.array([6_300, 6_480])    # размеры выборок
# Z-test
z_stat, p_value = proportions_ztest(counts, nobs, alternative='smaller')

print(z_stat)
print(f"{p_value:.20f}")
print(f"Группа A: {6300/9000:.2%}")
print(f"Группа B: {6750/9000:.2%}")
print(f"Абсолютная разница: {(6750-6300)/9000:.2%}")
print(f"Относительный прирост: {((75-70)/70)*100:.1f}%")
print(f"p-value: {p_value:.4f}")

if p_value < 0.05:
    print("✅ Результат статистически значим!")
else:
    print("❌ Разница не значима")
```
