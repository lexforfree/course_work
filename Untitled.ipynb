```python
import pandas as pd
import numpy as np
import glob
import os
from pathlib import Path


start_year = 2017

# adds dir
odds_path = './odds/' 
tml_path = './TML-Database/'
players_atp_path = "./TML-Database/ATP_Database.csv"
```

```python
excel_files = glob.glob(os.path.join(odds_path, '*.xlsx'))
excel_files =  list(filter(
    lambda f: int(Path(f).stem.strip()) >= start_year,
    excel_files
))
df_list = []

for file in excel_files:
    df = pd.read_excel(file)
    df_list.append(df)

odds_df = pd.concat(df_list, ignore_index=True)
odds_df = odds_df.replace('-', np.nan)
odds_df.sample(8).T
```

# Task 1

Цель: собрать признаки, которые объясняют исход матча, и сравнить с бейзлайном букмекеров (implied probability).


```python
import pandas as pd
import numpy as np

ods = pd.read_parquet('ods.parquet')
tml = pd.read_parquet('tml.parquet')
players = pd.read_parquet('players.parquet')

ods['Date'] = pd.to_datetime(ods['Date'])
tml['tourney_date'] = pd.to_datetime(tml['tourney_date'].astype(str), format='%Y%m%d', errors='coerce')

(ods.shape, tml.shape, players.shape)
```

```python
import re

def norm_name(s: str) -> str:
    if pd.isna(s):
        return ''
    s = str(s).strip().lower()
    s = re.sub(r"\s+", " ", s)
    s = s.replace("'", "")
    s = s.replace("-", " ")
    s = re.sub(r"[^a-z\.\s]", "", s)
    s = re.sub(r"\s+", " ", s).strip()
    return s

ods['winner_norm'] = ods['Winner'].map(norm_name)
ods['loser_norm'] = ods['Loser'].map(norm_name)
tml['winner_norm'] = tml['winner_name'].map(norm_name)
tml['loser_norm'] = tml['loser_name'].map(norm_name)

# tml.tourney_date = start of tournament week (typically Monday)
ods['tourney_date'] = (ods['Date'] - pd.to_timedelta(ods['Date'].dt.weekday, unit='D')).dt.normalize()
tml['tourney_date'] = tml['tourney_date'].dt.normalize()

ods['p1'] = ods[['winner_norm','loser_norm']].min(axis=1)
ods['p2'] = ods[['winner_norm','loser_norm']].max(axis=1)
tml['p1'] = tml[['winner_norm','loser_norm']].min(axis=1)
tml['p2'] = tml[['winner_norm','loser_norm']].max(axis=1)

ods_key = ods[['tourney_date','p1','p2']].copy()
tml_key = tml[['tourney_date','p1','p2']].copy()

(ods_key.isna().sum().sum(), tml_key.isna().sum().sum())
```

```python
# Join: try exact key on (tourney_date, p1, p2)
joined = (
    tml.merge(
        ods,
        how='inner',
        on=['tourney_date','p1','p2'],
        suffixes=('_tml','_ods')
    )
)

print('joined shape:', joined.shape)
print('match rate tml:', round(joined.shape[0] / tml.shape[0], 3))
print('match rate ods:', round(joined.shape[0] / ods.shape[0], 3))

joined[['tourney_date','winner_name','loser_name','Winner','Loser','surface','Surface','B365W','B365L','PSW','PSL']].head(5)
```
